```python
# calculating correlation intraday
import numpy as np

lags = np.arange(-1, 2, 1)
num_interv = int(len(dfs)/N_daily_data)
corr_list = np.zeros( (num_interv, 2) )
corrval = 0

for day in range(num_interv):
  start_idx = int(day*N_daily_data)
  end_idx = int((day+1)*N_daily_data)
  ts1 = merged_df.sent3stat[start_idx:end_idx]
  ts2 = merged_df.idx_change[start_idx:end_idx]
  rs = np.nan_to_num([ts1.corr(ts2.shift(lag)) for lag in lags])
  maxrs = np.max(rs)
  minrs = np.min(rs)
  if np.abs(maxrs) >= np.abs(minrs):
      corrval = maxrs
  else:
      corrval = minrs
  corr_list[day,0] = corrval 
  corr_list[day,1] = (lags[np.argmax(rs)])
corr_list
```
